```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

a = -0.85   # bifurcation parameter


def system(t, z):
    x, y = z

    dx = x + y
    dy = a*y + (a - 1)*x - x**3 - x**2*y

    return [dx, dy]


# time interval
t_span = (0, 50)
t_eval = np.linspace(0, 50, 4000)

# initial condition
z0 = [0.1, 0.1]

sol = solve_ivp(system, t_span, z0, t_eval=t_eval)

x = sol.y[0]
y = sol.y[1]

plt.figure(figsize=(6,6))
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Hopf System Phase Portrait")
plt.axis("equal")
plt.show()
```

```python
plt.figure(figsize=(6,6))

ics = [
    [0.1,0.1],
    [1,1],
    [-1,0],
    [-0.5,0.3],
    [0.2,-0.8],
]

for ic in ics:
    sol = solve_ivp(system,(0,50),ic,t_eval=np.linspace(0,50,2000))
    plt.plot(sol.y[0],sol.y[1])

plt.xlabel("x")
plt.ylabel("y")
plt.title("Hopf Bifurcation System")
plt.axis("equal")
plt.show()
```
